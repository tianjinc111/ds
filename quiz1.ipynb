# Part 1 (of 1):

Using the `df_car` DataFrame below, create a single plot (no subplots) which:
- scatters `model_year` & `mpg` on the x and y axes respectively
- only include a marker cars (rows) which have 4, 6 or 8 cylinder engines
- color code the markers for each car based on the number of cylinders
    - all 4 cylinder cars are represented with a distinct color
    - all 6 cylinder cars are represented with a distinct color
    - all 8 cylinder cars are represented with a distinct color
- plots a line representing the average `mpg` across all `model_year` for all cars with 4, 6 or 8 cylinders
    - note: line color should correspond to the scatter color chosen above
- give your graph all the appropriate labels (including a title) and manage any aesthetic / readability issues which come up
- print your final output graph to a pdf file (only submit the `.ipynb` file, not this `.pdf`)

Use matplotlib / seaborn (not plotly) to accomplish this task.

```python
import seaborn as sns

df_car = sns.load_dataset('mpg')
df_car.shape
```

```python
df_car.head()
```

```python
# importing all of the other needed packages here:
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# set sns here
sns.set()
```

```python
# assign the values here for each cars with 4,6,8 cylinders as well as calculated there mean.
four_cylinder = df_car.loc[df_car['cylinders'] == 4]
six_cylinder = df_car.loc[df_car['cylinders'] == 6]
eight_cylinder = df_car.loc[df_car['cylinders'] == 8]

four_cylinder_mean = four_cylinder['mpg'].mean()
six_cylinder_mean = six_cylinder['mpg'].mean()
eight_cylinder_mean = eight_cylinder['mpg'].mean()


# put the assigned values in the list for the latter ploting through for loop as well as create
# the color list and label lists.  
list1 = [four_cylinder,six_cylinder,eight_cylinder]
list2 = [four_cylinder_mean,six_cylinder_mean,eight_cylinder_mean]
colorList = ["red", "green", "blue"]
labelName_list = ["Car with 4 cylinder","Car with 6 cylinder","Car with 8 cylinder"]
labelAverage_list = ["Average of 4 cylinder","Average of 6 cylinder","Average of 8 cylinder"]


# ploting loop. 
for i in range(len(list1)):
    #each plot for 4,6,8 cylinders
    plt.scatter(list1[i]['model_year'],list1[i]['mpg'], alpha= 0.8,color = colorList[i],label = labelName_list[i])
    # each mean corresponding with thir cars, as well as the color are also the same. 
    plt.axhline(y = list2[i], color = colorList[i], label = labelAverage_list[i])



# set up the label, size, title
plt.legend()
plt.gcf().set_size_inches((13,7))
plt.suptitle("model_year and mpg for cars with different cylinders")
plt.xlabel("Years")
plt.ylabel("MPG")



# save the file into a pdf file
f_out = 'mpg_and_years.pdf'
with PdfPages(f_out) as pdf:
    pdf.savefig(bbox_inches='tight')
```
